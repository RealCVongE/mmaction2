```python
import cv2

def get_video_fps(video_path):
    # 비디오 파일 열기
    cap = cv2.VideoCapture(video_path)
    
    # 비디오 파일이 열렸는지 확인
    if not cap.isOpened():
        print("비디오 파일을 열 수 없습니다.")
        return None
    
    # FPS 가져오기
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    # 비디오 파일 닫기
    cap.release()
    
    return fps

# 비디오 파일 경로
video_path = 'C_3_12_1_BU_SMC_08-07_13-30-11_CC_RGB_DF2_M1.mp4'

# FPS 가져오기
fps = get_video_fps(video_path)

if fps is not None:
    print("비디오의 FPS는 {}입니다.".format(fps))
```

```python
import cv2

# 비디오 파일 경로
video_path = 'Sample_buying/Sample/01.원천데이터/02.구매행동/02.선택/C_2_2_10_BU_SYA_09-17_16-18-31_CA_RGB_DF1_F1_F1.mp4'

# 비디오 파일 열기
cap = cv2.VideoCapture(video_path)

# 누적 프레임 번호 변수
frame_number = 179
#마지막 1443
#1443-1080= 363
# 프레임 읽기
ret, frame = cap.read()

# 초기화
current_frame_number = cap.get(cv2.CAP_PROP_POS_FRAMES)  # 현재 프레임 번호
frame_time = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000  # 현재 프레임의 시간(초)

# 반복
while True:

    # 프레임 정보 표시
    cv2.putText(frame, f"Frame: {current_frame_number}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)
    cv2.putText(frame, f"Time: {frame_time:.2f} seconds", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)

    # 프레임 보여주기
    cv2.imshow('Frame', frame)

    # 키 입력 처리
    key = cv2.waitKey(1) & 0xFF

    # 왼쪽 화살표 키: 이전 프레임으로 이동
    if key == ord('a'):
        frame_number -= 1
        if frame_number < 0:
            frame_number = 0

    # 오른쪽 화살표 키: 다음 프레임으로 이동
    elif key == ord('d'):
        frame_number += 1

    # 'q' 키: 종료
    elif key == ord('q'):
        break

    # 프레임 설정 및 읽기
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    ret, frame = cap.read()

    # 프레임 정보 업데이트
    current_frame_number = cap.get(cv2.CAP_PROP_POS_FRAMES)  # 현재 프레임 번호
    frame_time = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000  # 현재 프레임의 시간(초)

# 비디오 해제
cap.release()

# 종료 메시지
print("종료되었습니다.")
```

```python
import xmltodict
import json

def convert_xml_to_json(xml_file_path, json_file_path):
    # XML 파일을 열고 읽기
    with open(xml_file_path, 'r', encoding='utf-8') as file:
        xml_string = file.read()
    
    # XML 문자열을 파이썬 딕셔너리로 변환
    dict_data = xmltodict.parse(xml_string)
    
    # 딕셔너리를 JSON 파일로 저장
    with open(json_file_path, 'w', encoding='utf-8') as json_file:
        json_file.write(json.dumps(dict_data, indent=4))

# 사용 예제
xml_file_path = 'Sample_buying/Sample/02.라벨링데이터/02.구매행동/02.선택/C_2_2_10_BU_SYA_09-17_16-18-31_CA_RGB_DF1_F1_F1.xml'
    # XML 파일을 열고 읽기
with open(xml_file_path, 'r', encoding='utf-8') as file:
    xml_string = file.read()

# XML 문자열을 파이썬 딕셔너리로 변환
dict_data = xmltodict.parse(xml_string)
print(len(dict_data['annotations']['meta']['task']['labels']['label']))

for i in dict_data['annotations']['meta']['task']['labels']['label']:
    print(i["name"])
print(dict_data['annotations']['track'][0].keys())
print(dict_data['annotations']['track'][0]['@label'])
print(len(dict_data['annotations']['track'][0]['box']))
print(dict_data['annotations']['track'][0]['box'][0].keys())
```
