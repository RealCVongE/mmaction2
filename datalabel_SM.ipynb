```python
# .mp4 확장자를 .pkl 확장자로 변경한 파일 경로들을 저장할 리스트
pkl_paths_updated = []

# 모든 파일 경로에 대해 확장자 변경 수행
def mp4_to_pkl(pkl_new_path):
    for path in pkl_new_path:
        # 파일 경로에서 .mp4를 .pkl로 변경
        new_path = path.replace('.mp4', '.pkl')
        # 변경된 경로를 새로운 리스트에 추가
        pkl_paths_updated.append(new_path)
    return pkl_paths_updated
```

```python
from tools.data.skeleton.ntu_pose_extraction_modify import ntu_pose_extraction
import mmengine

# args 변수를 사용하여 필요한 설정을 직접 정의
class Args:
    def __init__(self, device, det_config, det_checkpoint, det_score_thr, video, output, skip_postproc):
        self.device = device
        self.det_config = det_config
        self.det_checkpoint = det_checkpoint
        self.det_score_thr = det_score_thr
        self.video = video
        self.output = output
        self.skip_postproc = skip_postproc


# 필요한 인수를 설정
args = Args(
    device='cuda:0',
    det_config='demo/demo_configs/faster-rcnn_r50-caffe_fpn_ms-1x_coco-person.py',  # noqa: E501,
    det_checkpoint='https://download.openmmlab.com/mmdetection/v2.0/faster_rcnn/faster_rcnn_r50_fpn_1x_coco-person/faster_rcnn_r50_fpn_1x_coco-person_20201216_175929-d022e227.pth',  # noqa: E501
    det_score_thr=0.5,
    video= name[0],
    output= pkls[0],
    skip_postproc=False
)
```

```python
# 함수 호출
anno = ntu_pose_extraction(args.video, args.device, args.det_config, args.det_checkpoint, args.det_score_thr, args.skip_postproc)
```

```python
# 결과 저장
mmengine.dump(anno, args.output)
```
